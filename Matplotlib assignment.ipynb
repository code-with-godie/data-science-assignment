Matplotlib is a low level graph plotting library in python that serves as a visualization utility

### Example of a simple plot using matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
xPoints = np.array([0,20])
yPoints = np.array([0,255])
```

```python
plt.plot(xPoints,yPoints)
plt.show()
```

The plot() function is used to draw points (markers) in a diagram.

## Create Labels for a Plot

we use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

```python
plt.plot(xPoints,yPoints)
plt.xlabel('Student mean grade ')
plt.ylabel('Students marks')
plt.show()
```

## creating a title for the plot

use the title() function to set a title for the plot.

```python
plt.plot(xPoints,yPoints)
plt.xlabel('Student mean grade ')
plt.ylabel('Students marks')
plt.title('Student marks distribution in a class')
plt.show()
```

## Adding markers at each point

we use the marker parameter to add different marker style

```python
plt.plot(xPoints,yPoints,marker='*')
plt.show()
```

## Changing the line style

You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:

```python
plt.plot(xPoints,yPoints,linestyle = 'dashed', marker='d')
plt.show()
```

## Add Grid Lines to a Plot

 you can use the grid() function to add grid lines to the plot

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

## Display Multiple Plots

With the subplot() function you can draw multiple plots in one figure:

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

## Creating Scatter Plots

you can use the scatter() function to draw a scatter plot.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

## plotting two plots in a scalar plot

```python
#plot one
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#plot two
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

## Changing the color of each plot

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'purple')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#78c998')

plt.show()
```

## Color Each Dot

You can even set a specific color for each dot by using an array of colors as value for the c argument:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

## Creating Bars graphs

With Pyplot, you can use the bar() function to draw bar graphs:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

## Horizontal Bars

use the barh() function:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

## Bar Width

The bar() takes the keyword argument width to set the width of the bars:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```

## Histogram

we use the hist() function to create histograms.

## Creating Pie Charts

 use the pie() function to draw pie charts:

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

## Labels

Add labels to the pie chart with the labels parameter.

Start Angle

The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

## Shadow

Add a shadow to the pie chart by setting the shadows parameter to True:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, shadow = True)
plt.show() 
```

## Colors

The colors parameter, if specified, must be an array with one value for each wedge:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels,colors=mycolors)
```

## Legend

To add a list of explanation for each wedge, use the legend() function:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

## Legend With Header

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```
